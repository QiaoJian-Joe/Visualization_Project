```python
dict_count_tree = {}
```

```python
def count(col):
    if type(col.Genus) == type(0.01):
        col.Genus = 'UNKNOWN'
        col.Family = 'UNKNOWN'
    if type(col.Genus) == type('str') and len(col.Genus.strip())>1 and len(col.Family.strip())>1 and col['Year Planted'] in range(2005,2021):
        if col.Genus in dict_count_tree.keys():
            dict_count_tree[col.Genus]['value'] += 1
            dict_count_tree[col.Genus]['location'].append([col.Longitude,col.Latitude])
            if col['Year Planted'] in dict_count_tree[col.Genus]['years'].keys():
                dict_count_tree[col.Genus]['years'][col['Year Planted']]+=1
            else:
                dict_count_tree[col.Genus]['years'][col['Year Planted']] = 1
        else:
            dict_count_tree[col.Genus] = {}
            dict_count_tree[col.Genus]['value'] = 1
            dict_count_tree[col.Genus]['parent'] = col.Family
            dict_count_tree[col.Genus]['name'] = col.Genus
            dict_count_tree[col.Genus]['years'] = {}
            dict_count_tree[col.Genus]['location'] = [[col.Longitude,col.Latitude]]
```

```python
dataset.apply(count,axis = 1)
```

```python
tree_tree = {'children':[],'name':'Tree'}
final_count = {}
```

```python
for item in dict_count_tree:
    if dict_count_tree[item]['value'] > 200:
        final_count[item] = dict_count_tree[item]
```

```python
family_list = []
for item in final_count:
    family_list.append(dict_count_tree[item]['parent'])
```

```python
family_list = list(set(family_list))
```

```python
for tree in family_list:
    tree_tree['children'].append({'children':[],'name':tree,'colname':'level2'})
    for item in final_count:
        if final_count[item]['parent'] == tree and final_count[item]['value'] >= 200:
            tree_tree['children'][-1]['children'].append({'name':final_count[item]['name'],'value':final_count[item]['value'],'colname':'level3'})
```

```python
import json
json_str = json.dumps(tree_tree)
with open('test_data.json', 'w') as json_file:
    json_file.write(json_str)
```

```python
for item in final_count.values():
    for year in range(2005,2021):
        if year not in item['years'].keys():
            item['years'][year] = 0
            print(year,item)
```

```python
tmp_dict_pendding = {}
index = 0
for i in final_count:
    for j in final_count[i]['years'].keys():
        tmp_dict_pendding[index] = {'name':final_count[i]['name'],'year':j,'value':final_count[i]['years'][j],'parent':final_count[i]['parent'],'location':final_count[i]['location']}
        index += 1
    
```

```python
final_df = pd.DataFrame.from_dict(tmp_dict_pendding, orient='index')

final_df = final_df.reset_index(drop = True)

final_df = final_df.sort_values(by = 'year',axis = 0,ascending = True)

final_df.to_csv('count_tree.csv',index = False)
```

```python
value_df = final_df.sort_values(by = 'value',axis = 0,ascending = True)

value_df.to_csv('value_tree.csv',index = False)
```

```python
location_dict = {}
num = 0
for item in tmp_dict_pendding.values():
    for location in item['location']:
        location_dict[num] = {'location':location, 'name':item['name'],'parent':item['parent']}
        num += 1
```

```python
location_df = pd.DataFrame.from_dict(location_dict, orient='index')

location_df = location_df.reset_index(drop = True)

location_df.to_csv('location_tree.csv',index = False)
```
